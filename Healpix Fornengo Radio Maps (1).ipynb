```python
import numpy as np
import healpy as hp
# Set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
```

```python
def mapping(file,freq,Max,coords):
    dummylist=hp.read_map(file).tolist()
    for x in range(len(dummylist)):
        if dummylist[x]<=0:
            dummylist[x]=hp.UNSEEN
    if coords=="C":
        hp.mollview(np.array(dummylist),cmap="CMRmap",cbar=False,max=Max,
                min=0,title="%s Map" % freq,coord=("C","G"))
    if coords=="E":
        hp.mollview(np.array(dummylist),cmap="CMRmap",cbar=False,max=Max,
                min=0,title="%s Map" % freq,coord=("E","G"))
    if coords=="G":
        hp.mollview(np.array(dummylist),cmap="CMRmap",cbar=False,max=Max,
                min=0,title="%s Map" % freq,coord="G")
    hp.graticule()
```

```python
#fornengo files
mapping("lambda_DRAO_22MHz_1_256.fits","22 MHz",200000,"G")
mapping("MAIPU_MU_1_64.fits","45 MHz",30000,"G")
mapping("haslam408_dsds_Remazeilles2014.fits","408 MHz",100,"G")
mapping("Dwingeloo_Kelvins_1_256.fits", "820 MHz",30,"G")
mapping("STOCKERT+VILLA-ELISA_1420MHz_1_256.fits", "1420 MHz",12000,"G")
mapping("lambda_23de_hea.fits", "2326 MHz",1,"C")
```
